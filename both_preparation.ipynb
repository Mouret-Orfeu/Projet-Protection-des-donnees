```python
def physical_time_to_string(time,sec):
    d = datetime.strptime(time, '%d/%m/%Y %H:%M')
    
    string = f"{d.year}/{d.day}/{d.month} {d.hour}:{d.minute}:{sec}"

    return string

def network_time_to_string(string):

    try:
        d = datetime.strptime(string, '%Y-%m-%d %H:%M:%S.%f')
    except:
        d = datetime.strptime(string, '%Y-%m-%d %H:%M:%S')

    
    string = f"{d.year}/{d.day}/{d.month} {d.hour}:{d.minute}:{d.second}"

    return string
```

```python
def tout_changer_physical(df):
    serie = df["Time"].copy()
    premier = list(pd.unique(df["Time"]))[0]
    lim = len(df[df["Time"]==premier])
    prem_sec = 60-lim
    i=0
    for sec in range(prem_sec,60):
        serie[i]=physical_time_to_string(serie[i],sec)
        i+=1

    sec=0
    while i!=len(df):
        serie[i]=physical_time_to_string(serie[i],sec)
        i+=1
        sec+=1
        if sec==60:
            sec=0    
    ret_df = df.copy()
    ret_df["Time"] = serie

    return ret_df

def tout_changer_network(df):
    serie = df["Time"].copy()
    serie = serie.map(network_time_to_string)

    ret_df = df.copy()
    ret_df["Time"] = serie

    return ret_df
```
